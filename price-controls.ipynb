```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def demand_func(q):
    return -(q) + 100

supply = np.linspace(0, 100, 100)
demand = demand_func(supply)
```

```python
fig = plt.figure(figsize=(8,8), dpi= 80)
ax = fig.add_subplot(111)
plt.margins(x=0, y=0)
plt.ylabel("Price")
plt.xlabel("Quantity")
plt.plot(supply, supply, color="black", alpha=1)
plt.plot(supply, demand, color="black", alpha=1)
plt.text(15, 85, 'Demand', fontsize=15)
plt.text(73, 85, 'Supply', fontsize=15)

prices = [20, 50]
price_labels = ["$Price_{ceiling}$", "$Price_{no\ ceiling}$"]
quantity_labels = ["$Quantity_{ceiling}$", "$Quantity_{no\ ceiling}$"]
plt.yticks(prices, price_labels)
plt.xticks(prices, quantity_labels)

for p, l in zip(prices, price_labels):
    plt.hlines(p, xmin=0, xmax=demand_func(p), colors='black', alpha=1, linestyles="--", lw=1)

for p, l in zip(prices, price_labels):
    plt.vlines(p, ymin=0, ymax=demand_func(p), colors='black', alpha=1, linestyles="--", lw=1)

plt.fill_between(demand, supply, 50, where=(supply > 50)&(demand < 21), color='blue', alpha=0.5)
plt.fill_between(demand, 50, 20, where=(demand < 21), color='green', alpha=0.5)
plt.fill_between(supply, 20, supply, where=(demand>79), color='red', alpha=0.5)
plt.fill_between(supply, 50, demand, where=(demand<80)&(supply<50), color='yellow', alpha=0.5)
plt.fill_between(supply, 50, supply, where=(demand<80)&(supply<50), color='orange', alpha=0.5)

plt.text(8, 65, 'A', fontsize=15)
plt.text(8, 35, 'B', fontsize=15)
plt.text(8, 13, 'C', fontsize=15)
plt.text(24, 65, 'D', fontsize=15)
plt.text(24, 35, 'E', fontsize=15)

plt.savefig('price-control.svg', transparent=True)
plt.show()
```
